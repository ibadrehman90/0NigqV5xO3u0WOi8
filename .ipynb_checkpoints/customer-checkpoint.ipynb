```python
import pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold, cross_val_score
import numpy as np
from sklearn import tree
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from numpy import set_printoptions
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop('Y', axis=1)
y = df['Y']
```

```python
plt.hist(X, density=True, bins=6)  # density=False would make counts
plt.ylabel('Happy/Unhappy')
plt.xlabel('Responses');
```

```python
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
plt.boxplot(X)
 
# show plot
plt.show()
```

```python
plt.matshow(df.corr())
plt.show()
```

```python
scatter_matrix(df, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
```

```python
dt_tree = tree.DecisionTreeClassifier(criterion = 'entropy',random_state=0)
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scores = cross_val_score(dt_tree, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
rf = RandomForestClassifier(criterion='entropy',random_state=0)
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scores = cross_val_score(rf, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
lr = LogisticRegression()
cv = KFold(n_splits=10, random_state=1, shuffle=True)
scores = cross_val_score(lr, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
gnb = GaussianNB()
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scores = cross_val_score(gnb, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```
